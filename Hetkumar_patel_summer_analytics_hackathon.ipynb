```python
Y = X.iloc[:,-2:]
```

```python
X = X.iloc[:,:-2]
```

```python
X = pd.get_dummies(X,columns=['race','sex','marital_status','rent_or_own','employment_status','hhs_geo_region'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,
                                                    test_size=0.2)

X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter = 10000)
```

```python
lr1 = LogisticRegression(max_iter = 10000)
```

```python
Y_train
```

```python
lr.fit(X_train,Y_train['xyz_vaccine'])
```

```python
lr1.fit(X_train,Y_train['seasonal_vaccine'])
```

```python
Y_pred = lr.predict_proba(X_test)
```

```python
Y_pred1 = lr1.predict_proba(X_test)
```

```python
Y_pred
```

```python
Y_pred[:,1]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(Y_test['xyz_vaccine'],Y_pred[:,1],average='macro')
```

```python
roc_auc_score(Y_test['seasonal_vaccine'],Y_pred1[:,1],average='macro')
```
